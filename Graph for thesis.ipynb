```python
import pandas as pd

text1 = pd.read_csv('vgg1_record.csv')

text1 = text1.dropna(axis=0)

acc1 = list(text1['acc'])
val_acc1 = list(text1['val_acc'])

loss1 = list(text1['loss'])
val_loss1 = list(text1['val_loss'])
```

```python
text2 = pd.read_csv('vgg2_record.csv')

text2 = text2.dropna(axis=0)

acc2 = list(text2['acc'])
val_acc2 = list(text2['val_acc'])

loss2 = list(text2['loss'])
val_loss2 = list(text2['val_loss'])
```

```python
text3 = pd.read_csv('vgg3_record.csv')

text3 = text3.dropna(axis=0)

acc3 = list(text3['acc'])
val_acc3 = list(text3['val_acc'])

loss3 = list(text3['loss'])
val_loss3 = list(text3['val_loss'])
```

```python
import matplotlib.pyplot as plt

epochs = range(1, len(loss1)+1)

plt.figure(figsize=(6,4))
plt.plot(epochs,acc1,'r',linestyle = '-',label='Training accuracy')
plt.plot(epochs,val_acc1,'b',linestyle = '-',label='Validation accuracy')
plt.plot(epochs,acc2,'r',linestyle = ':')
plt.plot(epochs,val_acc2,'b',linestyle = ':')
plt.plot(epochs,acc3,'r',linestyle = '--')
plt.plot(epochs,val_acc3,'b',linestyle = '--')
plt.ylim([0.495, 1.005])
plt.title('Training and Validation accuracy')
plt.legend()

plt.show()

plt.figure(figsize=(6,4))
plt.plot(epochs, loss1, 'r', linestyle = '-', label='Training loss')
plt.plot(epochs, val_loss1, 'b', linestyle = '-', label='Validation loss')
plt.plot(epochs, loss2, 'r', linestyle = ':')
plt.plot(epochs, val_loss2, 'b', linestyle = ':')
plt.plot(epochs, loss3, 'r',linestyle = '--')
plt.plot(epochs, val_loss3, 'b', linestyle = '--')
plt.ylim([-0.05, 1.201])
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
